```
'''
tensorflow를 쓰지 않고 AndOperator 계산
for문 사용
'''
import numpy as np
import time
start = time.time()
def sigmoid(h):
  return 1/(1+np.exp(-h))

X = np.array([
              [0, 0],
              [0, 1],
              [1, 0],
              [1, 1]
], dtype = 'float32')

y = np.array([
              [0],
              [0],
              [0],
              [1]
], dtype = 'float32')
learning_rate = 0.1

w = np.array([[1], [2]], dtype = 'float32')
b = np.array([3], dtype = 'float32')

for i in range(1000):
  for j in range(2):
    hypothesis = sigmoid(np.dot(X, w) + b)
    print('hypothesis :\n', hypothesis)
    cost = -1/4 * (y * np.log(hypothesis) + (1 - y) * np.log(1 - hypothesis))
    print('cost :\n', cost)
    X_col_2d = X.T[j].reshape(1, -1)
    w_gred = 1/4 * np.sum(np.dot(X_col_2d, (hypothesis - y)))
    print('{}번째 시도 => w_gred : {}'.format(i, w_gred))
    w[j] = w[j] - learning_rate * w_gred
    print('w{} :{}'.format(j, w[j]))
    b_gred = 1/4 * np.sum(hypothesis - y)
    print('{}번째 시도 => b_gred : {}'.format(i, b_gred))
    b = b - learning_rate * b_gred
    print('b:{}'.format(b))

print('걸린 시간 : {:.0f}초'.format(time.time() - start))
```

```
'''
tensorflow를 쓰지 않고 AndOperator 계산
for문 사용
정확도까지 표시
'''
import numpy as np
import time
start = time.time()
def sigmoid(h):
  return 1/(1+np.exp(-h))

X = np.array([
              [0, 0],
              [0, 1],
              [1, 0],
              [1, 1]
], dtype = 'float32')

y = np.array([
              [0],
              [0],
              [0],
              [1]
], dtype = 'float32')
learning_rate = 0.1

w = np.array([[1], [2]], dtype = 'float32')
b = np.array([3], dtype = 'float32')

for i in range(1000):
  for j in range(2):
    hypothesis = sigmoid(np.dot(X, w) + b)
    cost = -1/4 * (y * np.log(hypothesis) + (1 - y) * np.log(1 - hypothesis))
    X_col_2d = X.T[j].reshape(1, -1)
    w_gred = 1/4 * np.sum(np.dot(X_col_2d, (hypothesis - y)))
    w[j] = w[j] - learning_rate * w_gred
    b_gred = 1/4 * np.sum(hypothesis - y)
    b = b - learning_rate * b_gred

print('걸린 시간 : {:.0f}초'.format(time.time() - start))

predict = sigmoid(np.dot(X, w) + b)
predict01 = np.where(predict >= 0.5, 1, 0)
predict02 = (predict01 == y)
acc = np.sum(predict02)/4
print('정확도 :', acc)
```
