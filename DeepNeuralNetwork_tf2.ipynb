```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```
X = np.array([
              [0, 0],
              [0, 1],
              [1, 0],
              [1, 1]
], dtype = 'float32')
y = np.array([0, 1, 1, 0], dtype = 'float32')
```

```
model = Sequential()
model.add(Dense(2, input_dim = 2, activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid')) # 그 전의 결과가 다음 input에 넘어감 => 순서대로 input이 넘어가기 때문에 Sequential 모델이다!
```

```
model.summary()
```

```
import time
start = time.time()
model.compile(loss = 'binary_crossentropy', optimizer = Adam(lr=0.01), metrics = ['acc'])
model.fit(X, y, epochs = 500)
print('걸린 시간 : {:.2f}초'.format(time.time() - start))
```

```
pred = model.predict(X)
pred
```

```
predict = (np.where(pred > 0.5, 1, 0).flatten() == y)
accuracy = np.sum(predict) / len(predict)
print('정확도 :', accuracy)
```

```
arr = np.array([[0, 1]], dtype = 'float32')
pred = np.where(model.predict(arr) > 0.5, 1, 0)
print('예측값 :', pred.flatten())
```

```
arr = np.array([[1, 1]], dtype = 'float32')
pred = np.where(model.predict(arr) > 0.5, 1, 0)
print('예측값 :', pred.flatten())
```
