```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
tf.enable_eager_execution()
tf.__version__
```

```
X = np.array([1, 2, 3], dtype = 'float32')
y = np.array([2, 2.5, 3.5], dtype = 'float32')
```

```
# X에 곱해지는 값 W를 선언, 초기값은 2
# tf.Variable :
#   Gradient Decent로 tf가 계산해줌
W = tf.Variable([2], dtype = 'float32')
b = tf.Variable([1], dtype = 'float32')
```

```
learning_rate = 0.1
```

```
with tf.GradientTape() as tape:
  hypothesis = W * X + b
  print('='*100)
  print('hypothesis :', hypothesis)
  print('(hypothesis - y) :', (hypothesis - y))
  print('tf.square(hypothesis - y) :', tf.square(hypothesis - y))
  print('tf.reduce_mean(tf.square(hypothesis - y)) :', tf.reduce_mean(tf.square(hypothesis - y)))
  cost = tf.reduce_mean(tf.square(hypothesis - y))
  print('cost :', cost)
  # print('tape.gradient(cost, [W,b]) :', tape.gradient(cost, [W,b])) # w 기울기
  w_gred, b_gred = tape.gradient(cost, [W,b]) # tape.gradient의 경우 딱 한번만 구해줌
  print('w_gred :', w_gred)
  print('b_gred :', b_gred)
  W.assign_sub(w_gred * learning_rate)
  b.assign_sub(b_gred * learning_rate)
  print('W :', W)
  print('b :', b)
```

```
for i in range(1000):
  with tf.GradientTape() as tape:
    hypothesis = W * X + b
    print('='*100)
    print('hypothesis :', hypothesis)
    print('(hypothesis - y) :', (hypothesis - y))
    print('tf.square(hypothesis - y) :', tf.square(hypothesis - y))
    print('tf.reduce_mean(tf.square(hypothesis - y)) :', tf.reduce_mean(tf.square(hypothesis - y)))
    cost = tf.reduce_mean(tf.square(hypothesis - y))
    print('cost :', cost)
    # print('tape.gradient(cost, [W,b]) :', tape.gradient(cost, [W,b])) # w 기울기
    w_gred, b_gred = tape.gradient(cost, [W,b]) # tape.gradient의 경우 딱 한번만 구해줌
    print('w_gred :', w_gred)
    print('b_gred :', b_gred)
    W.assign_sub(w_gred * learning_rate)
    b.assign_sub(b_gred * learning_rate)
    print('W :', W)
    print('b :', b)
```

```
%tensorflow_version 1.X
import tensorflow as tf
import numpy as np
X = np.array([1,2,3], dtype = 'float32')
y = np.array([2, 2.5, 3.5], dtype = 'float32')
W = tf.Variable([2], dtype = 'float32')
b = tf.Variable([1], dtype = 'float32')
learning_rate = 0.1
```

```
for i in range(1000):
  with tf.GradientTape() as tape:
    hypothesis = W * X + b
    cost = tf.reduce_mean(tf.square(hypothesis - y))
  
  w_grad, b_grad = tape.gradient(cost, [W, b])
  W.assign_sub(w_grad * learning_rate)
  b.assign_sub(b_grad * learning_rate)
```

```
print('i = {}, w = {}, b = {}, cost = {}'.format(i, W.numpy(), b.numpy(), cost))
```

```
import matplotlib.pyplot as plt
plt.plot(X, y, marker = 'o')
plt.plot(X, hypothesis.numpy())
plt.ylim(0, 4)
```
