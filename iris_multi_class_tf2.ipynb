```
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```
data = datasets.load_iris()
X = data.data
y = data.target
y
```

```
X
```

```
y = y.reshape(-1, 1)
y
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_train
```

```
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(y)
enc.transform(y_train).toarray()
```

```
y_train_onehot = np.array(enc.transform(y_train).toarray(), dtype = 'float32')
y_test_onehot = np.array(enc.transform(y_test).toarray(), dtype = 'float32')
```

```
model = Sequential()
model.add(Dense(16, input_dim = 4, activation = 'sigmoid'))
model.add(Dense(3, activation='softmax'))
```

```
model.summary()
```

```
model.compile(loss = 'categorical_crossentropy', optimizer = Adam(lr = 0.1), metrics= ['acc'])
model.fit(X_train, y_train_onehot, epochs = 1000)
```

```
X_test = scaler.transform(X_test) 
# hypothesis0 = tf.sigmoid(tf.matmul(X_test, w0) + b0)
# predict = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
predict = model.predict(X_test)
# predict   
predict01 = np.argmax(predict, 1)
target = np.argmax(y_test_onehot, 1)
predict02 = (predict01 == target)
accuracy = np.sum(predict02)/len(predict02)
print('')
print(accuracy)
```
